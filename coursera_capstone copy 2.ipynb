```python
import pandas as pd
import numpy as np
```

```python
#importing beautifulsoup for extraction
from bs4 import BeautifulSoup
import requests
```

```python
#creating the soup
r = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
content = r.text
soup = BeautifulSoup(content, 'html.parser')    
```

```python
#lookup for the table in the html
table=soup.table
```

```python
#initializing all the columns and look for the rows in the table
postcode=[]
borough=[]
neighbourhood=[]
rows=table.find_all('tr')
```

```python
#for each row, assing the variable to each array
for row in rows:
    try:
        postcode.append(str(row.find_all('td')[0].text))
        borough.append(str(row.find_all('td')[1].text))
        s=str(row.find_all('td')[2].text)
        neighbourhood.append(s.replace('\n',''))
    
    #there is an error on the first values are the are table headers and not table data
    except:
        postcode.append('Postcode')
        borough.append('Borough')
        neighbourhood.append('Neighbourhood')
```

```python
#initialize values for a new panda dataframe
headers = ('PostalCode', 'Borough', 'Neighbourhood')
data = (postcode, borough, neighbourhood)
data_dict = dict(zip(headers, data))

#initialize the dataframe
df1 = pd.DataFrame(data_dict)

#drop the first line as it the headers
df1.drop([0],axis=0,inplace=True)

#set index Borough so that we can delete all the Not assigned values from it
df1.set_index('Borough',drop=True,inplace=True)
df1.drop('Not assigned',axis=0,inplace=True)

#reset index so that we can see the table as it is and reorganize it
df1.reset_index(inplace=True)
df1=df1[["PostalCode","Borough","Neighbourhood"]]
```

```python
df1
```

```python
#Modify all the Neighbourhoods that are "Not assigned" and return them to the table.
df1['Neighbourhood']=np.where(df1['Neighbourhood']=="Not assigned",df1['Borough'],df1['Neighbourhood'])
df1
```

```python
#Unify all the postcodes by their neighbourhood
df2=df1.groupby('PostalCode')['Neighbourhood'].apply(lambda x: ', '.join(x)).reset_index()
#Merge results of the groupby with the borough information
df3=pd.merge(df2,df1[['PostalCode','Borough']], on='PostalCode')
#Clean all the duplicates
df3.drop_duplicates(subset=['Neighbourhood','Borough'], keep='first',inplace=True)
#reset the index and reorganize the table.
df3.reset_index(inplace=True, drop=True)
df3=df3[["PostalCode","Borough","Neighbourhood"]]
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            100)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=df4['Neighbourhood'],
                                   latitudes=df4['Latitude'],
                                   longitudes=df4['Longitude']
                                  )
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
tornto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
#grouped by neighbourhood
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
toronto_grouped.shape
```

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#transforming it into a pandas dataframe
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```
